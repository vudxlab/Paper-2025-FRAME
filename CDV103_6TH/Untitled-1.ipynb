```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# Load file .mat
mat_file = scipy.io.loadmat('DATA/SETUP1_TH1.mat')

# Lấy danh sách các key bắt đầu bằng "Untitled"
untitled_keys = [key for key in mat_file.keys() if key.startswith("Untitled")]

data_matrices = []

for key in untitled_keys:
    try:
        data = mat_file[key][0, 0][0]  # Lấy dữ liệu thực tế
        arr = data.flatten()  # Chuyển thành mảng 1D
        if arr.size == 0:
            print(f"{key} is empty, skipping.")
        else:
            data_matrices.append(arr)
            print(f"{key}: shape {arr.shape}")
    except Exception as e:
        print(f"Lỗi khi xử lý {key}: {e}")

if data_matrices:
    # Kiểm tra xem các mảng có độ dài giống nhau không
    lengths = [arr.size for arr in data_matrices]
    print("Array lengths:", lengths)
    
    # Nếu tất cả có cùng độ dài, có thể dùng vstack
    if len(set(lengths)) == 1:
        merged_matrix = np.vstack(data_matrices)
    else:
        print("Các mảng có độ dài khác nhau. Không thể ghép nối trực tiếp.")
        merged_matrix = None

    if merged_matrix is not None:
        print("\nMa trận 2D sau khi ghép nối:")
        print(merged_matrix.shape)


# Vẽ đồ thị
DATA = merged_matrix[2:,:1000]
plt.figure(figsize=(10, 5))
for row in DATA:
    plt.plot(row, alpha=0.6)
plt.xlabel('Chỉ số mẫu')
plt.ylabel('Giá trị')
plt.title('Dữ liệu sau khi ghép thành ma trận 2 chiều')
plt.show()
```

```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

# Danh sách tên file (giả sử các file nằm trong thư mục DATA)
files = [f"DATA/SETUP1_TH{i}.mat" for i in range(1, 9)]
data_3d_list = []

for file in files:
    print(f"Đang xử lý file: {file}")
    try:
        mat_file = scipy.io.loadmat(file)
    except Exception as e:
        print(f"Lỗi khi load {file}: {e}")
        continue

    # Lấy danh sách các key bắt đầu bằng "Untitled"
    untitled_keys = [key for key in mat_file.keys() if key.startswith("Untitled")]
    
    data_matrices = []
    
    for key in untitled_keys:
        try:
            # Giả sử dữ liệu thực nằm ở [0,0][0] của mỗi key
            data = mat_file[key][0, 0][0]
            arr = data.flatten()  # Chuyển thành mảng 1D
            if arr.size == 0:
                print(f"{key} trong {file} trống, bỏ qua.")
            else:
                data_matrices.append(arr)
                print(f"{key} trong {file}: shape {arr.shape}")
        except Exception as e:
            print(f"Lỗi khi xử lý {key} trong {file}: {e}")
    
    if data_matrices:
        # Kiểm tra xem các mảng có độ dài giống nhau không
        lengths = [arr.size for arr in data_matrices]
        if len(set(lengths)) == 1:
            # Ghép các mảng theo chiều dọc
            merged_matrix = np.vstack(data_matrices)
            # Bỏ 2 hàng đầu và chỉ lấy 1000 cột đầu (nếu có đủ cột)
            if merged_matrix.shape[1] >= 1000000:
                DATA = merged_matrix[1:, 400000:600000]
            else:
                print(f"Warning: ma trận trong {file} có ít hơn 1000 cột, lấy tất cả các cột.")
                DATA = merged_matrix[1:, :]
            data_3d_list.append(DATA)
            print(f"{file}: DATA shape {DATA.shape}")
        else:
            print(f"Các mảng trong {file} có độ dài khác nhau. Bỏ qua file này.")
    else:
        print(f"Không có dữ liệu hợp lệ trong {file}.")

# Ghép các ma trận 2 chiều thành ma trận 3 chiều (chiều thứ nhất ứng với các file)
if data_3d_list:
    data_3d = np.stack(data_3d_list, axis=0)
    print("\nMa trận 3 chiều sau khi ghép nối:", data_3d.shape)
else:
    print("Không có dữ liệu để tạo ma trận 3 chiều.")

# Ví dụ: Vẽ dữ liệu của file đầu tiên trong ma trận 3 chiều (mỗi dòng của DATA là 1 đường plot)
if data_3d_list:
    plt.figure(figsize=(10, 5))
    for i, row in enumerate(data_3d[0]):
        plt.plot(row, alpha=0.6)
    plt.xlabel('Chỉ số mẫu')
    plt.ylabel('Giá trị')
    plt.title('Đồ thị dữ liệu của file đầu tiên')
    plt.show()
```
